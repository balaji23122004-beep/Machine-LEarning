```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target

model = MLPClassifier(hidden_layer_sizes=(5,),
                      max_iter=1000)
model.fit(X, y)

print("Training Accuracy:", model.score(X, y))
```
